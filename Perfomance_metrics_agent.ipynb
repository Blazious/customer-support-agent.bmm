```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set(style="whitegrid", palette="pastel")
%matplotlib inline
```

# üìä SureLife AI Agent Performance Report

This notebook visualizes the key performance metrics of the SureLife AI customer support agent.  
We will cover:

1. Overall Accuracy
2. Strengths Accuracy
3. Customer Service Readiness
4. Deployment Phase Metrics

All visualizations are designed to highlight areas of strength and areas requiring improvement.


```python
# -------------------------------
# Step 2: Overall Performance Bar Chart
# -------------------------------

categories = ['Correct Answers', 'Incorrect Answers', 'Missed Information']
values = [58, 8, 8]
colors = ['#4CAF50', '#F44336', '#FF9800']

plt.figure(figsize=(8,5))
bars = plt.bar(categories, values, color=colors)
plt.title('SureLife AI Agent - Overall Performance', fontsize=16)
plt.ylabel('Number of Questions', fontsize=12)

# Add data labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', fontsize=11)

plt.show()
```

**Interpretation:**

- ‚úÖ Correct Answers: 58/74  
- ‚ùå Incorrect Answers: 8/74  
- ‚ö† Missed Information: 8/74  

The agent correctly answered ~78% of questions but has room for improvement in missed information handling.


```python
# -------------------------------
# Step 3: Strengths Pie Chart
# -------------------------------

strengths = ['Knowledge Retrieval', 'Complex Information Synthesis', 'Appropriate "I don\'t know"']
accuracy = [85, 80, 100]
colors = ['#2196F3','#FFC107','#4CAF50']

plt.figure(figsize=(7,7))
plt.pie(accuracy, labels=strengths, autopct='%1.1f%%', startangle=140, colors=colors, shadow=True)
plt.title('Strengths Accuracy of SureLife AI Agent', fontsize=16)
plt.show()
```

**Interpretation:**

- Knowledge Retrieval is strong at 85%  
- Complex Synthesis at ~80%  
- Appropriately admitting "I don't know" 100% of the time  

The pie chart visually emphasizes that while factual retrieval is strong, synthesis still has slight room for improvement.


```python
# -------------------------------
# Step 4: Customer Service Readiness Bar Chart
# -------------------------------

metrics = ['Reliability', 'Empathy', 'Proactive Service', 'Communication Style']
scores = [75, 40, 30, 50]
colors = sns.color_palette("pastel", len(metrics))

plt.figure(figsize=(10,6))
bars = plt.barh(metrics, scores, color=colors)
plt.xlabel('Performance (%)', fontsize=12)
plt.title('Customer Service Readiness Metrics', fontsize=16)
plt.xlim(0,100)

# Data labels
for i, v in enumerate(scores):
    plt.text(v + 2, i, f"{v}%", va='center', fontsize=11)

plt.show()
```

**Interpretation:**

- Reliability is decent at 75%  
- Empathy and Proactive Service are low (40% and 30%)  
- Communication Style is robotic (50%)  

These highlight key areas to improve before deployment to ensure customer satisfaction.


```python
# -------------------------------
# Step 5: Deployment Phase Metrics Line Chart
# -------------------------------

phases = ['Phase 1', 'Phase 2', 'Phase 3']
query_resolution = [70, 85, 95]
response_accuracy = [75, 90, 98]

plt.figure(figsize=(8,5))
plt.plot(phases, query_resolution, marker='o', color='#4CAF50', label='Query Resolution Rate (%)')
plt.plot(phases, response_accuracy, marker='o', color='#2196F3', label='Response Accuracy (%)')
plt.title('Deployment Phase Metrics', fontsize=16)
plt.ylabel('Percentage (%)', fontsize=12)
plt.ylim(0,100)
plt.legend()
plt.grid(True)
plt.show()
```

**Interpretation:**

- Query resolution and response accuracy improve as the system progresses through deployment phases  
- Phase 3 shows near-target performance (~95-98%), indicating readiness for autonomous operation once critical fixes are applied.


```python
# -------------------------------
# Step 6: Optional - Combined Dashboard
# -------------------------------

# For a single dashboard, we can use subplots
fig, axs = plt.subplots(2, 2, figsize=(15,12))

# Overall Performance Bar
axs[0,0].bar(categories, values, color=colors)
axs[0,0].set_title('Overall Performance')
axs[0,0].set_ylabel('Number of Questions')

# Strengths Pie
axs[0,1].pie(accuracy, labels=strengths, autopct='%1.1f%%', startangle=140, colors=colors, shadow=True)
axs[0,1].set_title('Strengths Accuracy')

# Customer Service Readiness
axs[1,0].barh(metrics, scores, color=colors)
axs[1,0].set_xlabel('Performance (%)')
axs[1,0].set_title('Customer Service Readiness')
axs[1,0].set_xlim(0,100)

# Deployment Phase Line
axs[1,1].plot(phases, query_resolution, marker='o', color='#4CAF50', label='Query Resolution Rate')
axs[1,1].plot(phases, response_accuracy, marker='o', color='#2196F3', label='Response Accuracy')
axs[1,1].set_title('Deployment Phase Metrics')
axs[1,1].set_ylim(0,100)
axs[1,1].legend()
axs[1,1].grid(True)

plt.tight_layout()
plt.show()
```

**Dashboard Summary:**

- Combines all key metrics in a single view  
- Quick visual insights into strengths, weaknesses, and readiness for deployment  
- Helps prioritize areas for improvement before launch
